```python
# importing pandas and matplotlib
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# loading data into pandas dataframe 
df = pd.read_csv('DataScienceSalaries.csv')

df.head(15)
```

```python
unnecessaryColumns = ['salary', 'salary_currency']

df.drop(columns = unnecessaryColumns, inplace = True) # removing unnecessary columns from the dataframe

df.columns # displaying the name of columns in the dataframe 
```

```python
df.dropna(inplace = True) # Removing NaN missing values
```

```python
duplicateValues = df.duplicated() # checking for duplicate values in the dataframe

totalDuplicateValues = duplicateValues.sum() # counting the total duplicate values

print('Total Duplicate Values: ', totalDuplicateValues)
print('\n')
print('Duplicate Values: \n', duplicateValues.sample(15))
```

```python
# displaying unique values in all the columns
for column in df.columns: # iterating over each column in the dataframe
    if df[column].dtype not in ['int64', 'float64']: # verifying if the values in the column are not of int64 or float64 datatype
        uniqueValues =  df[column].unique() # selecting only unique values from the columns
        print(column, ': ', uniqueValues, '\n')
```

```python
renamingValues = {"SE": "Senior Level/Expert",
                    "MI": "Medium Level/Intermediate",
                    "EN": "Entry Level",
                    "EX": "Executive Level"}

df['experience_level'].replace(renamingValues, inplace=True) # renaming the values in experience_level column

df.sample(15)
```

```python
def columnInput():
    '''
        purpose: Take column input from the user and then validate that input
        outcome: Valid column name
    '''

    # Take input from the user for the column name
    column = input("Enter the column name: ")

    # Check if the input is a valid column name in the DataFrame
    if column in df.columns:
        # Check if the column contains numerical values
        if df[column].dtype in ['int64', 'float64']:
            return column # if column is valid and contains numerical values
        else:
            # if column doesn't contains numerical values
            print("Column contains non-numeric values. Please enter a column with numeric values") 
            return -1
    else:
        # if input for the column name is invalid
        print("Please enter a valid column name")
        return -1
        
```

```python
def sumOfColumn(column):
    '''
    purpose: Calculate the sum of the provided column.
    input: column - list containing the values from particular column
    output: Sum of the column
    '''

    # Initialize the sum to 0
    sum_ = 0

    # Iterate through each value in the column and add it to the sum
    for i in column:
        sum_ += i
    
    # Return the sum of the column
    return sum_
```

```python
column = columnInput() # calling columnInput() function to take column name as input and validate it

if column != -1:  # Check if the column input is valid
    print(f"Sum of {column} column (user-defined function): ", sumOfColumn(df[column]))

    print(f"Sum of {column} column (in-built function): ", df[column].sum())
```

```python
def meanOfColumn(column):
    '''
    purpose: Calculate the mean of the given column.
    input: column - list containing the values from particular column
    output: Mean of the column
    '''

    # Calculate the sum of the column
    sum_ = sumOfColumn(column)

    # Calculate the total number of values in the column
    totalNumber = len(column)

    # Calculate the mean of the column
    mean = sum_ / totalNumber

    # Return the mean of the column
    return mean
```

```python
column = columnInput() # calling columnInput() function to take column name as input and validate it

if column != -1:  # Check if the column input is valid
    print(f"Mean of {column} column (user-defined function): ", meanOfColumn(df[column]))

    print(f"Mean of {column} column (in-built function): ", df[column].mean())
```

```python
def standardDeviation(column):
    '''
    purpose: Calculate the standard deviation of the given column.
    input: column - list containing the values from particular column
    output: Standard deviation of the column
    '''

    # Initialize a variable that stores the value of numerator portion of the standard deviation formula
    numeratorPortion = 0

    # Calculate the total number of values in the column
    totalLines = len(column) - 1 
    
    # Iterate through each value in the column to calculate the numerator portion
    for i in column: 
        numeratorPortion += ((i - meanOfColumn(column)) ** 2)
    
    # Calculate the standard deviation
    standardDeviation = (numeratorPortion / totalLines) ** 0.5
    
    # Return the standard deviation of the column
    return standardDeviation
```

```python
column = columnInput() # calling columnInput() function to take column name as input and validate it

if column != -1:  # Check if the column input is valid
    print(f"Standard Deviation of {column} column (user-defined function): ", standardDeviation(df[column]))

    print(f"Standard Deviation of {column} column (in-built function): ", df[column].std())
```

```python
def skewness(column):
    '''
    purpose: Calculate the skewness of the given column.
    input: column - list containing the values from particular column
    output: Skewness of the column
    '''

    # Initialize a variable that stores the value of numerator portion of the skewness formula
    numeratorPortion = 0
    
    # Calculate the total number of values in the column
    totalLines = len(column)
    
    # Iterate through each value in the column to calculate the numerator portion
    for i in column:
        numeratorPortion += ((i - meanOfColumn(column)) ** 3)
    
    # Calculate the value of denominator portion of the skewness formula
    denominatorPortion = (totalLines - 1) * (standardDeviation(column) ** 3)
    
    # Calculate the skewness 
    skewness = numeratorPortion / denominatorPortion
    
    # Return the skewness of the column
    return skewness
```

```python
column = columnInput() # calling columnInput() function to take column name as input and validate it

if column != -1:  # Check if the column input is valid
    print(f"Skewness of {column} column (user-defined function): ", skewness(df[column]))

    print(f"Skewness of {column} column (in-built function): ", df[column].skew())
```

```python
def kurtosis(column):
    '''
    purpose: Calculate the kurtosis of the given column.
    input: column - list containing the values from particular column
    output: Kurtosis of the column
    '''

    # Initialize a variable that stores the value of numerator portion of the kurtosis formula
    numeratorPortion = 0
    
    # Calculate the total number of values in the column
    totalLines = len(column)
    
    # Iterate through each value in the column to calculate the numerator portion
    for i in column:
        numeratorPortion += ((i - meanOfColumn(column)) ** 4)
    
    # Calculate the denominator portion of the kurtosis formula
    denominatorPortion = (totalLines - 1) * (standardDeviation(column) ** 4)
    
    # Calculate the kurtosis
    kurtosis = (numeratorPortion / denominatorPortion) - 3
    
    # Return the kurtosis of the column
    return kurtosis
```

```python
column = columnInput() # calling columnInput() function to take column name as input and validate it

if column != -1:  # Check if the column input is valid
    print(f"Kurtosis of {column} column (user-defined function): ", kurtosis(df[column]))

    print(f"Kurtosis of {column} column (in-built function): ", df[column].kurt())
```

```python
def correlation(column1, column2):
    '''
    purpose: Calculate the correlation coefficient between two columns.
    input:
        column1 - list containing the values from first column
        column2 - list containing the values from second column
    output: Correlation coefficient between the two columns
    '''

    # Calculate the mean of each column
    mean1 = meanOfColumn(column1)
    mean2 = meanOfColumn(column2)

    # Initialize variables for the numerator and denominators of the correlation formula
    numerator = 0
    denominator1 = 0
    denominator2 = 0
    
    # Iterate through the values of both columns to calculate the numerator and denominators
    for i in range(len(column1)):
        numerator += (column1[i] - mean1) * (column2[i] - mean2)
        denominator1 += (column1[i] - mean1) ** 2
        denominator2 += (column2[i] - mean2) ** 2

    # Calculate the correlation coefficient 
    correlation = numerator / ((denominator1 * denominator2) ** 0.5)
    
    # Return the correlation coefficient
    return correlation
```

```python
# calculate the correlation between the different numerical columns in the dataframe
row1 = [correlation(df['work_year'], df['work_year']), correlation(df['work_year'], df['salary_in_usd']), correlation(df['work_year'], df['remote_ratio'])]
row2  = [correlation(df['salary_in_usd'], df['work_year']), correlation(df['salary_in_usd'], df['salary_in_usd']), correlation(df['salary_in_usd'], df['remote_ratio'])]
row3 = [correlation(df['remote_ratio'], df['work_year']), correlation(df['remote_ratio'], df['salary_in_usd']), correlation(df['remote_ratio'], df['remote_ratio'])]

# Store correlation values in a list
correlationValues = [row1, row2, row3]

# Define column names for the resulting DataFrame
columns = ['work_year', 'salary_in_usd', 'remote_ratio']

# Create a DataFrame using the correlation values, with columns and index specified
correlationDf = pd.DataFrame(correlationValues, columns=columns, index=columns)

# Display the correlation DataFrame
correlationDf
```

```python
# finding out top 15 jobs
top15Jobs = df['job_title'].value_counts().head(15)

top15Jobs
```

```python
# Extracting titles of top 15 jobs
jobTitles = top15Jobs.index 

# Extracting frequency of top 15 jobs
jobCount = top15Jobs.values 

# Plotting bar graph for top 15 jobs
plt.figure(figsize=(12, 7)) # defines the size of bar graph
plt.bar(jobTitles, jobCount, label='Frequency', color="red", alpha=0.5) # defines the style of bar graph
plt.legend() # includes the label
plt.title('Top 15 Jobs', fontweight='bold') # defines the title of bar graph
plt.xlabel('Job Title', fontweight='bold') # defines label for x-axis
plt.ylabel('Job Count', fontweight='bold') # defines label for y-axis
plt.xticks(rotation=45, ha='right') # defines custom marker for x-axis
plt.yticks(range(0, 1101, 100)) # defines custom marker for y-axis
plt.show() # displays the plot
```

```python
# Calculating salaries based on job titles
salaries = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(15)

print(salaries)

# Extracting job titles and average salaries
jobTitles = salaries.index
averageSalaries = salaries.values

# Plotting bar graph for salaries based on job title
plt.figure(figsize=(30, 8))  # Defines the size of the plot
plt.bar(jobTitles, averageSalaries, color='blue', alpha=0.6)  # Plotting the bar graph
plt.title('Salaries based on Job Title', fontweight='bold', fontsize=24)  # Defines the title of the plot
plt.xlabel('Job Title', fontweight='bold', fontsize=18)  # Defines label for the x-axis
plt.ylabel('Average Salary (USD)', fontweight='bold', fontsize=18)  # Defines label for the y-axis
plt.xticks(rotation=90, ha='center', fontsize=14)  # Rotates and aligns x-axis labels
plt.yticks(range(0, 400000, 25000), fontsize=18)  # Sets the y-axis ticks
plt.show()  # Displays the plot
```

```python
# Calculating salaries based on experience levels
salariesByExperience = df.groupby('experience_level')['salary_in_usd'].mean()

print(salariesByExperience)

# Extracting experience levels and average salaries
experienceLevel = salariesByExperience.index
averageSalariesByExperience = salariesByExperience.values

# Plotting bar graph for salaries based on experience level
plt.figure(figsize=(12, 6)) # Defines the size of the plot
plt.bar(experienceLevel, averageSalariesByExperience, color='red', alpha=0.5) # plotting the bar graph
plt.title('Salaries based on Experience Level', fontweight='bold')  # Defines the title of the plot
plt.xlabel('Experience Level', fontweight='bold')  # Defines label for the x-axis
plt.ylabel('Average Salary (USD)', fontweight='bold')  # Defines label for the y-axis
plt.xticks(ha='center')  # Defines the alignment of x-axis labels
plt.show()  # Displays the plot
```
